```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = './1950_through_2020.csv'
df = pd.read_csv(file_path)
pd.set_option('display.max_rows', None)

df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%b')
df.set_index('Date', inplace=True)

def graph_funct(line1, color1):
    
        yearly_temps = df.resample('YE')[line1].mean()

        plt.plot(yearly_temps.index.year, yearly_temps, marker='o',color=color1)

        plt.xlabel('Year')
        plt.ylabel('Temperature (F)')
        plt.title(f'Yearly {line1} Temperatures NYC(LGA)')
        plt.grid(True)

        plt.show()

graph_funct('Average','blue')
```
